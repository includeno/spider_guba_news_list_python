# 统计已抓取的新闻中评论个数非0的新闻 ✅

```python
#统计csv文件夹内所有csv 已抓取的新闻中评论个数非0的新闻个数，并展示总评论数
import os
import pandas as pd
csv_folder_path = "csv"
csv_files=os.listdir(csv_folder_path)
total_comment_count=0
total_count=0
print("count of csv_files:",len(csv_files))
total_df=None
for csv_file in csv_files:
    # print 文件名按_分隔中的stock信息
    if csv_file.startswith('links_guba_') and csv_file.count('_')==3:
        #print(csv_file.split('_')[2])
        df=pd.read_csv(os.path.join(csv_folder_path,csv_file))
        if(total_df is None):
            total_df=df
        else:
            total_df=pd.concat([total_df, df])
        total_count+=df.count()[0]
        total_comment_count+=df[df['comment_count']>0].shape[0]
print(total_count)
print(total_comment_count)

temp=total_df.sort_values(by=['comment_count'],ascending=False)
df_selected = temp[['comment_count']].head(10000)
count=0
for index, row in df_selected.iterrows():
    #print(row['comment_count'])
    count=count+row['comment_count']
print(count)
```
